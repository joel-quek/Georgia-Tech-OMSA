```r
# Assuming you have loaded the Carseats dataset into a data frame called "Carseats"

# Create the dummy variable for "Bad_Shelf"
Carseats$Bad_Shelf <- ifelse(Carseats$ShelveLoc == "Bad", 1, 0)

# Create the dummy variable for "Good_Shelf"
Carseats$Good_Shelf <- ifelse(Carseats$ShelveLoc == "Good", 1, 0)
```

```r
PriceDemand = read.csv("PriceDemand.csv", header = TRUE)
```

```r
head(PriceDemand)
```

```r
model3 <- lm(Qty~Price, data=PriceDemand)
summary(model3)
```

```r
# Assuming you have a dataset called "data" with columns "Price" and "Qty"

# Create the new variable Price_ln by taking the natural logarithm of Price
PriceDemand$Price_ln <- log(PriceDemand$Price)

# Fit the linear-log model
model4 <- lm(Qty ~ Price_ln, data = PriceDemand)

# Print the model summary
summary(model4)
```

```r
# Assuming you have a dataset called "data" with columns "Price" and "Qty"

# Fit the log-linear model
model5 <- lm(log(Qty) ~ Price, data = PriceDemand)

# Print the model summary
summary(model5)
```

```r
# Assuming you have a dataset called "data" with columns "Price" and "Qty"

# Fit the log-log model
model6 <- lm(log(Qty) ~ log(Price), data = PriceDemand)

# Print the model summary
summary(model6)
```
